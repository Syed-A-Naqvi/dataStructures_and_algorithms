```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(['A','B','C','D','E','F','H'])
G.add_weighted_edges_from([('A','C',6),('A','B',4),('C','B',8),('C','F',2),('C','D',14),('D','F',5),
                           ('B','E',10),('E','F',7),('E','H',4),('E','G',2),('G','H',1),('F','H',2)])
```

```python
def find_set(nodeSets, element):
    for s in nodeSets:
        if element in s:
            return s
    return None

def mst_Kruskal(G):
    
    # empty graph
    A = nx.Graph()
    
    # list of nodes sets
    nodeSets=[]
    for v in G.nodes():
        nodeSets.append(set(v))
    
    # edges sorted in ascending order by weight
    edges = []
    for v1, v2, w in G.edges(data=True):
        edges.append((v1, v2, w['weight']))
    edges.sort(key=(lambda x: x[2]))
    
    print(nodeSets)
    for e in edges:    
        set1 = find_set(nodeSets, e[0])
        set2 = find_set(nodeSets, e[1])
        if set1 != set2:
            A.add_weighted_edges_from([e])
            # Merge the two sets and update the nodeSets list
            merged_set = set1.union(set2)
            nodeSets.remove(set1)
            nodeSets.remove(set2)
            nodeSets.append(merged_set)
            print(nodeSets)
    
    return A
```

```python
pos1 = nx.spring_layout(G)  # positions for all nodes
# nodes
nx.draw_networkx_nodes(G, pos1, node_size=300)
# edges
nx.draw_networkx_edges(G, pos1)
# labels
nx.draw_networkx_labels(G, pos1, font_size=14, font_family='sans-serif')
edge_labels1 = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos1, edge_labels=edge_labels1)
plt.show()

A = mst_Kruskal(G)

pos2 = nx.spring_layout(A)  # positions for all nodes
# nodes
nx.draw_networkx_nodes(A, pos2, node_size=300)
# edges
nx.draw_networkx_edges(A, pos2)
# labels
nx.draw_networkx_labels(A, pos2, font_size=14, font_family='sans-serif')
edge_labels2 = nx.get_edge_attributes(A, 'weight')
nx.draw_networkx_edge_labels(A, pos2, edge_labels=edge_labels2)
plt.show()
```
